# Basic Time Crystal Simulation

This notebook demonstrates the basic usage of the time crystal simulation framework using the Floquet kicked-Ising model.

## Implementation of Equation (1) from the paper:

The Floquet kicked-Ising propagator is:

$$U_F = \exp\left[-\frac{i\tau}{2}\sum_j\left(J\sigma_j^{z}\sigma_{j+1}^{z}+h\sigma_j^{z}\right)\right]$$
$$\quad\times\exp\left[-\frac{i\pi}{2}\sum_j\sigma_j^{x}\right]$$
$$\quad\times\exp\left[-\frac{i\tau}{2}\sum_j\left(J\sigma_j^{z}\sigma_{j+1}^{z}+h\sigma_j^{z}\right)\right]$$

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
import sys
sys.path.append('../src')

from core.tensor_utils import create_initial_state
from core.observables import calculate_loschmidt_echo, magnetization
from models.kicked_ising import KickedIsingModel
from dynamics.tebd_evolution import CustomFloquet

# Set up plotting
plt.style.use('seaborn-v0_8')
plt.rcParams['figure.figsize'] = (10, 6)
```

```python
# System parameters
n_sites = 12
J = 1.0
h_disorder = 0.3 * J
tau = 1.0

# Create the kicked-Ising model
model = KickedIsingModel(
    n_sites=n_sites,
    J=J,
    h_disorder=h_disorder,
    tau=tau,
    disorder_seed=42
)

print(f"System initialized with {n_sites} sites")
print(f"J = {J}, h/J = {h_disorder/J}, tau = {tau}")

# Create initial state
psi_initial = create_initial_state(n_sites, state_type="neel")
print(f"Initial state created: {psi_initial}")
print(f"Initial magnetization: {magnetization(psi_initial, 'z'):.3f}")
```
